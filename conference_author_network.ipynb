```python
import time
import csv
import networkx as nx
from networkx.algorithms import bipartite
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import matplotlib.font_manager as fm
```

```python
font_path = '/Library/Fonts/NanumGothic.ttf'
font_name = fm.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
edgelist = []
f1 = csv.reader(open('conference_author.csv','r'))
for row in f1:
    edgelist.append((row[0],row[1]))
print(edgelist)
```

```python
Conference = []
Name= []
for i in edgelist:
    print(i[0] + ', ' + i[1])
    Conference.append(i[0])
    Name.append(i[1])
```

```python
B = nx.Graph()
B.add_nodes_from(Name, bipartite=0)
B.add_nodes_from(Conference, bipartite=1)
B.add_edges_from(edgelist)
```

```python
projected = nx.projected_graph(B, Name)
options = {
     'node_color': 'blue',
     'edge_color': 'red',
     'width' : 5
 }
# plt.savefig("Graph.png", dpi=300, format="PNG")
plt.figure(1,figsize=(10,10), dpi=300) 
plt.axis('off')
# fig = nx.draw(projected, with_labels=True, **options)
pos=nx.fruchterman_reingold_layout(projected, k=0.5)
# nx.draw(projected)
nx.draw_networkx_labels(projected, pos, label_pos = 0.3, font_weight = "bold", font_family='AppleGothic', font_size=8)
nx.draw_networkx_nodes(projected, pos, node_color="yellow")
nx.draw_networkx_edges(projected, pos, edge_color="green")
```
